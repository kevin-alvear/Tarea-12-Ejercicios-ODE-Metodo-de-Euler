1. **Usa el método de Euler** para aproximar las soluciones de los siguientes problemas de valor inicial:

   a. $y' = te^{3t-2y}, \quad 0 \leq t \leq 1, \quad y(0) = 0, \quad \text{con } h = 0.5$  
   b. $y' = 1 + (t-y)^2, \quad 2 \leq t \leq 3, \quad y(2) = 1, \quad \text{con } h = 0.5$  
   c. $y' = 1 + \frac{y}{t}, \quad 1 \leq t \leq 2, \quad y(1) = 2, \quad \text{con } h = 0.25$  
   d. $y' = \cos 2t + \sin 3t, \quad 0 \leq t \leq 1, \quad y(0) = 1, \quad \text{con } h = 0.25$

Este script resuelve los cuatro problemas del ejercicio 1:

**Problemas:**

**a.** $y' = t \cdot e^{3t} - 2y$, $0 \leq t \leq 1$, $y(0) = 0$, con $h = 0.5$  
**b.** $y' = 1 + (t - y)^2$, $2 \leq t \leq 3$, $y(2) = 1$, con $h = 0.5$  
**c.** $y' = 1 + \frac{y}{t}$, $1 \leq t \leq 2$, $y(1) = 2$, con $h = 0.25$  
**d.** $y' = \cos(2t) + \sin(3t)$, $0 \leq t \leq 1$, $y(0) = 1$, con $h = 0.25$

**El código hace lo siguiente para cada problema:**

1. Aplica el método de Euler.
2. Imprime los resultados para cada paso.
3. Grafica la solución aproximada.

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_method(f, y0, t0, tf, h):
    t = np.arange(t0, tf + h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(1, len(t)):
        y[i] = y[i-1] + h * f(t[i-1], y[i-1])

    return t, y

# Funciones para cada problema
def f1(t, y):
    return t * np.exp(3*t) - 2*y

def f2(t, y):
    return 1 + (t - y)**2

def f3(t, y):
    return 1 + y/t

def f4(t, y):
    return np.cos(2*t) + np.sin(3*t)

# Parámetros para cada problema
problems = [
    (f1, 0, 0, 1, 0.5),
    (f2, 1, 2, 3, 0.5),
    (f3, 2, 1, 2, 0.25),
    (f4, 1, 0, 1, 0.25)
]

# Resolvemos cada problema
for i, (f, y0, t0, tf, h) in enumerate(problems, 1):
    t, y = euler_method(f, y0, t0, tf, h)

    print(f"\nProblema {i}:")
    for j in range(len(t)):
        print(f"t = {t[j]:.2f}, y = {y[j]:.6f}")

    plt.figure(figsize=(10, 6))
    plt.plot(t, y, 'bo-', label=f'Aproximación de Euler')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Solución aproximada para el problema {i}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

   b. $y(1.25)$ y $y(1.93)$

   d. $y(0.54)$ y $y(0.94)$

Este script hace lo siguiente para cada problema:

**Pasos:**

1. Aplica el método de Euler para obtener la solución aproximada.
2. Para cada punto de interpolación especificado:
    - Encuentra los puntos más cercanos en la solución de Euler.
    - Realiza una interpolación lineal para estimar $y(t)$ en el punto deseado.
    - Calcula el valor real usando la función exacta.
    - Calcula el error absoluto entre el valor interpolado y el valor real.
3. Imprime los resultados para cada punto de interpolación.

**Puntos de interpolación para cada problema:**

**a.** $y(0.25)$ y $y(0.93)$  
**b.** $y(1.25)$ y $y(1.93)$  
**c.** $y(2.10)$ y $y(2.75)$  
**d.** $y(0.54)$ y $y(0.94)$

**Notas:**

Esto te permitirá ver qué tan precisas son las aproximaciones obtenidas mediante interpolación lineal de los resultados del método de Euler, en comparación con los valores reales de las funciones.

```python
import numpy as np

def euler_method(f, y0, t0, tf, h):
    t = np.arange(t0, tf + h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(1, len(t)):
        y[i] = y[i-1] + h * f(t[i-1], y[i-1])

    return t, y

def linear_interpolation(t, t0, t1, y0, y1):
    return y0 + (y1 - y0) * (t - t0) / (t1 - t0)

# Funciones para cada problema del ejercicio 3
def f1(t, y):
    return y/t - (y/t)**2

def f2(t, y):
    return 1 + y/t + (y/t)**2

def f3(t, y):
    return -(y + 1)*(y + 3)

def f4(t, y):
    return -5*y + 5*t**2 + 2*t

# Soluciones reales del ejercicio 4
def real_sol1(t):
    return t / (1 + np.log(t))

def real_sol2(t):
    return t * np.tan(np.log(t))

def real_sol3(t):
    return -3 + 2 / (1 + np.exp(-2*t))

def real_sol4(t):
    return t**2 + (1/3) * np.exp(-5*t)

# Parámetros para cada problema
problems = [
    (f1, real_sol1, 1, 1, 2, 0.1, [0.25, 0.93]),
    (f2, real_sol2, 0, 1, 3, 0.2, [1.25, 1.93]),
    (f3, real_sol3, -2, 0, 2, 0.2, [2.10, 2.75]),
    (f4, real_sol4, 1/3, 0, 1, 0.1, [0.54, 0.94])
]

for i, (f, real_f, y0, t0, tf, h, interpolate_points) in enumerate(problems, 1):
    t, y = euler_method(f, y0, t0, tf, h)

    print(f"\nProblema {i}:")
    for point in interpolate_points:
        # Encontrar los índices de los puntos más cercanos
        idx = np.searchsorted(t, point)
        if idx == 0:
            idx = 1
        elif idx == len(t):
            idx = len(t) - 1

        # Interpolar
        y_interp = linear_interpolation(point, t[idx-1], t[idx], y[idx-1], y[idx])

        # Calcular el valor real
        y_real = real_f(point)

        print(f"Para t = {point}:")
        print(f"  Valor interpolado: {y_interp:.6f}")
        print(f"  Valor real: {y_real:.6f}")
        print(f"  Error absoluto: {abs(y_real - y_interp):.6f}")
```

6. **Usa el método de Taylor de orden 2** para aproximar las soluciones de los siguientes problemas de valor inicial:

   a. $y' = te^{3t-2y}, \quad 0 \leq t \leq 1, \quad y(0) = 0, \quad \text{con } h = 0.5$  
   b. $y' = 1 + (t-y)^2, \quad 2 \leq t \leq 3, \quad y(2) = 1, \quad \text{con } h = 0.5$  
   c. $y' = 1 + \frac{y}{t}, \quad 1 \leq t \leq 2, \quad y(1) = 2, \quad \text{con } h = 0.25$  
   d. $y' = \cos 2t + \sin 3t, \quad 0 \leq t \leq 1, \quad y(0) = 1, \quad \text{con } h = 0.25$

Este script resuelve los cuatro problemas del ejercicio 6 (que son los mismos del ejercicio 1) usando el método de Taylor de orden 2:

**Problemas:**

**a.** $y' = t \cdot e^{3t} - 2y$, $0 \leq t \leq 1$, $y(0) = 0$, con $h = 0.5$  
**b.** $y' = 1 + (t - y)^2$, $2 \leq t \leq 3$, $y(2) = 1$, con $h = 0.5$  
**c.** $y' = 1 + \frac{y}{t}$, $1 \leq t \leq 2$, $y(1) = 2$, con $h = 0.25$  
**d.** $y' = \cos(2t) + \sin(3t)$, $0 \leq t \leq 1$, $y(0) = 1$, con $h = 0.25$

**El código hace lo siguiente para cada problema:**

1. Define la función $f(t, y)$ y sus derivadas parciales $\frac{\partial f}{\partial t}$ y $\frac{\partial f}{\partial y}$.
2. Aplica el método de Taylor de orden 2.
3. Imprime una tabla con los valores de $t$ y la aproximación de $y$ para cada paso.
4. Grafica la solución aproximada.

**Notas:**

Este método de Taylor de orden 2 debería proporcionar aproximaciones más precisas que el método de Euler usado en el ejercicio 1, especialmente para pasos $h$ más grandes.

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_order2_method(f, df_dt, df_dy, y0, t0, tf, h):
    t = np.arange(t0, tf + h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(1, len(t)):
        ti = t[i-1]
        yi = y[i-1]
        y[i] = yi + h * f(ti, yi) + (h**2 / 2) * (df_dt(ti, yi) + df_dy(ti, yi) * f(ti, yi))

    return t, y

# Funciones para cada problema y sus derivadas parciales
def f1(t, y):
    return t * np.exp(3*t) - 2*y

def df1_dt(t, y):
    return np.exp(3*t) + 3*t*np.exp(3*t)

def df1_dy(t, y):
    return -2

def f2(t, y):
    return 1 + (t - y)**2

def df2_dt(t, y):
    return 2*(t - y)

def df2_dy(t, y):
    return -2*(t - y)

def f3(t, y):
    return 1 + y/t

def df3_dt(t, y):
    return -y/t**2

def df3_dy(t, y):
    return 1/t

def f4(t, y):
    return np.cos(2*t) + np.sin(3*t)

def df4_dt(t, y):
    return -2*np.sin(2*t) + 3*np.cos(3*t)

def df4_dy(t, y):
    return 0

# Parámetros para cada problema
problems = [
    (f1, df1_dt, df1_dy, 0, 0, 1, 0.5),
    (f2, df2_dt, df2_dy, 1, 2, 3, 0.5),
    (f3, df3_dt, df3_dy, 2, 1, 2, 0.25),
    (f4, df4_dt, df4_dy, 1, 0, 1, 0.25)
]

# Resolvemos cada problema
for i, (f, df_dt, df_dy, y0, t0, tf, h) in enumerate(problems, 1):
    t, y = taylor_order2_method(f, df_dt, df_dy, y0, t0, tf, h)

    print(f"\nProblema {i}:")
    print("   t        y")
    print("------------------")
    for j in range(len(t)):
        print(f"{t[j]:6.2f}  {y[j]:11.6f}")

    plt.figure(figsize=(10, 6))
    plt.plot(t, y, 'bo-', label='Aproximación de Taylor orden 2')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Solución aproximada para el problema {i}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def taylor_order4_method(f, df_dt, df_dy, d2f_dt2, d2f_dtdy, d2f_dy2, d3f_dt3, d3f_dt2dy, d3f_dtdy2, d3f_dy3, y0, t0, tf, h):
    t = np.arange(t0, tf + h, h)
    y = np.zeros(len(t))
    y[0] = y0

    for i in range(1, len(t)):
        ti = t[i-1]
        yi = y[i-1]
        f_val = f(ti, yi)
        df_dt_val = df_dt(ti, yi)
        df_dy_val = df_dy(ti, yi)

        term1 = h * f_val
        term2 = (h**2 / 2) * (df_dt_val + df_dy_val * f_val)
        term3 = (h**3 / 6) * (d2f_dt2(ti, yi) + 2*d2f_dtdy(ti, yi)*f_val + d2f_dy2(ti, yi)*f_val**2 + df_dy_val*(df_dt_val + df_dy_val*f_val))
        term4 = (h**4 / 24) * (d3f_dt3(ti, yi) + 3*d3f_dt2dy(ti, yi)*f_val + 3*d3f_dtdy2(ti, yi)*f_val**2 + d3f_dy3(ti, yi)*f_val**3 +
                               3*d2f_dtdy(ti, yi)*(df_dt_val + df_dy_val*f_val) + d2f_dy2(ti, yi)*f_val*(3*df_dt_val + df_dy_val*f_val) +
                               df_dy_val*(d2f_dt2(ti, yi) + 2*d2f_dtdy(ti, yi)*f_val + d2f_dy2(ti, yi)*f_val**2 + df_dy_val*(df_dt_val + df_dy_val*f_val)))

        y[i] = yi + term1 + term2 + term3 + term4

    return t, y

# Definiciones para el problema 1
def f1(t, y):
    return t * np.exp(3*t) - 2*y

def df1_dt(t, y):
    return np.exp(3*t) + 3*t*np.exp(3*t)

def df1_dy(t, y):
    return -2

def d2f1_dt2(t, y):
    return 6*np.exp(3*t) + 9*t*np.exp(3*t)

def d2f1_dtdy(t, y):
    return 0

def d2f1_dy2(t, y):
    return 0

def d3f1_dt3(t, y):
    return 27*np.exp(3*t) + 27*t*np.exp(3*t)

def d3f1_dt2dy(t, y):
    return 0

def d3f1_dtdy2(t, y):
    return 0

def d3f1_dy3(t, y):
    return 0

# Definiciones para el problema 2
def f2(t, y):
    return 1 + (t - y)**2

def df2_dt(t, y):
    return 2*(t - y)

def df2_dy(t, y):
    return -2*(t - y)

def d2f2_dt2(t, y):
    return 2

def d2f2_dtdy(t, y):
    return -2

def d2f2_dy2(t, y):
    return 2

def d3f2_dt3(t, y):
    return 0

def d3f2_dt2dy(t, y):
    return 0

def d3f2_dtdy2(t, y):
    return 0

def d3f2_dy3(t, y):
    return 0

# Definiciones para el problema 3
def f3(t, y):
    return 1 + y/t

def df3_dt(t, y):
    return -y/t**2

def df3_dy(t, y):
    return 1/t

def d2f3_dt2(t, y):
    return 2*y/t**3

def d2f3_dtdy(t, y):
    return -1/t**2

def d2f3_dy2(t, y):
    return 0

def d3f3_dt3(t, y):
    return -6*y/t**4

def d3f3_dt2dy(t, y):
    return 2/t**3

def d3f3_dtdy2(t, y):
    return 0

def d3f3_dy3(t, y):
    return 0

# Definiciones para el problema 4
def f4(t, y):
    return np.cos(2*t) + np.sin(3*t)

def df4_dt(t, y):
    return -2*np.sin(2*t) + 3*np.cos(3*t)

def df4_dy(t, y):
    return 0

def d2f4_dt2(t, y):
    return -4*np.cos(2*t) - 9*np.sin(3*t)

def d2f4_dtdy(t, y):
    return 0

def d2f4_dy2(t, y):
    return 0

def d3f4_dt3(t, y):
    return 8*np.sin(2*t) - 27*np.cos(3*t)

def d3f4_dt2dy(t, y):
    return 0

def d3f4_dtdy2(t, y):
    return 0

def d3f4_dy3(t, y):
    return 0

# Parámetros para cada problema
problems = [
    (f1, df1_dt, df1_dy, d2f1_dt2, d2f1_dtdy, d2f1_dy2, d3f1_dt3, d3f1_dt2dy, d3f1_dtdy2, d3f1_dy3, 0, 0, 1, 0.5),
    (f2, df2_dt, df2_dy, d2f2_dt2, d2f2_dtdy, d2f2_dy2, d3f2_dt3, d3f2_dt2dy, d3f2_dtdy2, d3f2_dy3, 1, 2, 3, 0.5),
    (f3, df3_dt, df3_dy, d2f3_dt2, d2f3_dtdy, d2f3_dy2, d3f3_dt3, d3f3_dt2dy, d3f3_dtdy2, d3f3_dy3, 2, 1, 2, 0.25),
    (f4, df4_dt, df4_dy, d2f4_dt2, d2f4_dtdy, d2f4_dy2, d3f4_dt3, d3f4_dt2dy, d3f4_dtdy2, d3f4_dy3, 1, 0, 1, 0.25)
]

# Resolvemos cada problema
for i, (f, df_dt, df_dy, d2f_dt2, d2f_dtdy, d2f_dy2, d3f_dt3, d3f_dt2dy, d3f_dtdy2, d3f_dy3, y0, t0, tf, h) in enumerate(problems, 1):
    t, y = taylor_order4_method(f, df_dt, df_dy, d2f_dt2, d2f_dtdy, d2f_dy2, d3f_dt3, d3f_dt2dy, d3f_dtdy2, d3f_dy3, y0, t0, tf, h)

    print(f"\nProblema {i}:")
    print("   t        y")
    print("------------------")
    for j in range(len(t)):
        print(f"{t[j]:6.2f}  {y[j]:11.6f}")

    plt.figure(figsize=(10, 6))
    plt.plot(t, y, 'bo-', label='Aproximación de Taylor orden 4')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.title(f'Solución aproximada para el problema {i}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
